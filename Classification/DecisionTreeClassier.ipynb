## Separate data into X and Y features

```python
X = df.drop(columns='Drug')
Y = df[["Drug"]]
```

```python
X.head()
```

```python
Y.head()
```

## Data Preprocessing and Data Cleaning

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
# separate X data into categorical and continuous
cat = list(X.columns[X.dtypes=="object"])
con=list(X.columns[X.dtypes!="object"])
```

```python
print(cat)
```

```python
print(con)
```

```python
cat_pipe = make_pipeline(
    (SimpleImputer(strategy='most_frequent')),
    (OneHotEncoder(handle_unknown='ignore',sparse_output=False))
)
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy='mean'),
    StandardScaler()
)
```

```python
pre = ColumnTransformer([
    ("cat",cat_pipe,cat),
    ("con",num_pipe,con)
]).set_output(transform='pandas')
```

```python
pre
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

## Split the data into training and testing

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X_pre,Y,train_size=0.75,random_state=42)
```

```python
xtrain.head()
```

```python
xtest.head()
```

```python
ytrain.head()
```

```python
ytest.head()
```

## Build the Decision Tree CLassifier model

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(
    max_depth=5,
    min_samples_split=5,
    min_samples_leaf=2,
    criterion='gini'
)
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
from sklearn.model_selection import GridSearchCV
base_model = DecisionTreeClassifier()
```

max_depth=5,
    min_samples_split=5,
    min_samples_leaf=2,
    criterion='gini'

```python
params ={
    'max_depth' : [2,3,4,5,6,7,8],
    'min_samples_split':[1,2,3,4,6,8,10,12],
    'min_samples_leaf':[1,2,4,6,8,10,11],
    'criterion':['gini','entropy']
}
```

```python
gscv = GridSearchCV(base_model,param_grid=params,cv=4,scoring="f1_macro")
gscv.fit(xtrain,ytrain)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
best_dtc = gscv.best_estimator_
best_dtc
```

```python
best_dtc.score(xtrain,ytrain)
```

```python
best_dtc.score(xtest,ytest)
```

## Confusion Matrix

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(best_dtc,xtest,ytest)
```

```python
ypred = best_dtc.predict(xtest)
ypred[:5]
```

```python
ytest.head()
```

```python
yprobs = best_dtc.predict_proba(xtest)
yprobs[:5]
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred))
```

## Above model is good model as it is resulting in f1_macro avg >0.8
